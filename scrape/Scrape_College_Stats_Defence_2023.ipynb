```python
#upload combine_data and draft_data

import pandas as pd

# Read combine data from CSV
combine_data = pd.read_csv(r"C:\Users\benfi\Defensive_Players_NFL\data\combine_table.csv")

# Read draft data from CSV
draft_data = pd.read_csv(r"C:\Users\benfi\Defensive_Players_NFL\data\draft_table.csv")

year=2023
```

```python
# Get all unique values in the 'combine_pos' column of the combine_data DataFrame
unique_combine_pos = combine_data['combine_pos'].unique()

print("Unique values in combine_pos column:")
print(unique_combine_pos)
```

```python
def_combine = combine_data[(combine_data['combine_pos'].isin(['OLB', 'CB', 'S', 'ILB', 
                                                              'DE', 'DT', 'DL', 'EDGE', 'DB', 'SAF']) &
                           (combine_data['combine_Year'] == year))].copy()

def_combine.head()
```

```python
unique_draft_pos = draft_data['draft_pos'].unique()

print("Unique values in draft_pos column:")
print(unique_draft_pos)
```

```python
def_draft = draft_data[(draft_data['draft_pos'].isin(['OLB', 'CB', 'S', 'ILB', 
                                                            'DE', 'DT', 'DL', 'DB', 'LB'])&
                           (draft_data['draft_Year'] == year))].copy()




def_draft.head()
```

```python
combine_url = def_combine['combine_college_stats_url'].dropna()

draft_url = def_draft['draft_college_url'].dropna()

all_urls = pd.concat([combine_url, draft_url])

all_urls = all_urls.dropna()

all_urls = all_urls.drop_duplicates()

all_urls = [url for url in all_urls if url != '0']
```

```python
all_urls
```

```python
num_urls = len(all_urls)
print("Number of urls: ", num_urls)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# List to store DataFrames
dataframes = []

# Iterate through URLs
for url in all_urls:
    try:
        # Send a GET request to the URL
        response = requests.get(url)
        
        # Check if the request was successful
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            # Find the table with id 'rushing'
            table = soup.find('table', {'id': 'defense'})
            
            # Check if the table exists
            if table:
                # Extract data from the table
                data = []
                rows = table.find_all('tr')
                for row in rows[1:]:
                    data.append([cell.text.strip() for cell in row.find_all(['th', 'td'])])
                
                # Create a DataFrame
                df = pd.DataFrame(data)
                
                # Add URL column and populate it with the current URL
                df['URL'] = url
                
                # Append the DataFrame to the list
                dataframes.append(df)
            else:
                print(f"No table found for {url}")
        else:
            print(f"Failed to fetch {url}")
    except Exception as e:
        print(f"Error scraping {url}: {e}")

    # Introduce a time delay of 4 seconds per iteration
    time.sleep(4)

# Concatenate all DataFrames into a single DataFrame
if dataframes:
    combined_df = pd.concat(dataframes, ignore_index=True)
    print(combined_df)
else:
    print("No data to display.")
```

```python
combined_df_filtered = combined_df
# Define the new column names
new_column_names = {
    0: 'Year',
    1: 'School',
    2: 'Conf',
    3: 'Class',
    4: 'Pos',
    5: 'G',
    6: 'college_solo_tackles',
    7: 'college_assisted_tackle',
    8: 'college_total_tackles',
    9: 'college_tackles_for_loss',
    10: 'college_sk',
    11: 'college_int',
    12: 'college_int_yds',
    13: 'college_avg_int_yds',
    14: 'college_int_td',
    15: 'college_passes_defended',
    16: 'college_fumbles_recov',
    17: 'college_fumbles_recov_yds',
    18: 'college_fumbles_recov_td',
    19: 'college_fumbles_forced',
}

# Rename the columns
combined_df_filtered = combined_df_filtered.rename(columns=new_column_names)
```

```python
combined_df_filtered.dtypes
```

```python
combined_df_filtered['Year'] = combined_df_filtered['Year'].str.replace('*', '')

# Show the resulting DataFrame
print(combined_df_filtered)
```

```python
# Define the columns to convert to numeric
columns_to_convert = ['Year','G', 'college_solo_tackles', 'college_assisted_tackle', 'college_total_tackles',
                      'college_tackles_for_loss', 'college_sk', 'college_int', 'college_int_yds', 
                      'college_avg_int_yds', 'college_int_td', 'college_passes_defended', 
                      'college_fumbles_recov', 'college_fumbles_recov_yds', 'college_fumbles_recov_td', 
                      'college_fumbles_forced']

# Iterate over each column and convert to numeric
for col in columns_to_convert:
    # Replace any non-numeric values with NaN and coerce to numeric
    combined_df_filtered[col] = pd.to_numeric(combined_df_filtered[col], errors='coerce')

# Verify the conversion
print(combined_df_filtered.dtypes)
```

```python
# Drop rows where 'Year' column contains values 'Year', 'Career', or is blank
filtered_df = combined_df_filtered[~combined_df_filtered['Year'].isin(['Year', 'Career', ''])]

# Show the resulting DataFrame
print(filtered_df)
```

```python
# Drop rows where 'Year' column contains NaN
cleaned_df = combined_df_filtered.dropna(subset=['Year'])

cleaned_df.drop('college_avg_int_yds', axis=1, inplace=True)

# Show the resulting DataFrame
print(cleaned_df)
```

```python
import numpy as np

# Group by 'URL' and apply aggregation functions
agg_functions = {
    'Year': lambda x: '',  # Return empty string for 'Year'
    'School': lambda x: x.mode().iloc[0],  # Return mode of 'School'
    'Conf': lambda x: x.mode().iloc[0],  # Return mode of 'Conf'
    'Class': lambda x: '',  # Return empty string for 'Class'
    'Pos': lambda x: '',  # Return empty string for 'Pos'
    'G': 'sum',  # Sum of 'G'
    'college_solo_tackles': 'sum',  # Sum of 'college_solo_tackles'
    'college_assisted_tackle': 'sum',  # Sum of 'college_assisted_tackle'
    'college_total_tackles': 'sum',  # Sum of 'college_total_tackles'
    'college_tackles_for_loss': 'sum',  # Sum of 'college_tackles_for_loss'
    'college_sk': 'sum',  # Sum of 'college_sk'
    'college_int': 'sum',  # Sum of 'college_int'
    'college_int_yds': 'sum',  # Sum of 'college_int_yds'
    'college_int_td': 'sum',  # Sum of 'college_int_td'
    'college_passes_defended': 'sum',  # Sum of 'college_passes_defended'
    'college_fumbles_recov': 'sum',  # Sum of 'college_fumbles_recov'
    'college_fumbles_recov_yds': 'sum',  # Sum of 'college_fumbles_recov_yds'
    'college_fumbles_recov_td': 'sum',  # Sum of 'college_fumbles_recov_td'
    'college_fumbles_forced': 'sum'  # Sum of 'college_fumbles_forced'
}

# Apply aggregation
grouped_df = cleaned_df.groupby('URL').agg(agg_functions).reset_index()
```
