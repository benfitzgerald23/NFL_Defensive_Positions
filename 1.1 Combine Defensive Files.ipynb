```python
import os
import pandas as pd
import numpy as np
```

```python


# Get the current working directory
current_dir = os.getcwd()


# Specify the directory containing the CSV files
csv_dir = os.path.join(current_dir, 'data')


# List all files in the specified directory
files_in_dir = os.listdir(csv_dir)

# Filter out only the files that start with "def"
def_files = [file for file in files_in_dir if file.startswith('def')]
print(f'Filtered files: {def_files}')

```

```python
# List to hold the dataframes
dataframes = []

# Loop through the list of "def" files and read them into dataframes
for file in def_files:
    file_path = os.path.join(csv_dir, file)
    df = pd.read_csv(file_path)
    dataframes.append(df)

# Concatenate all dataframes
combined_df = pd.concat(dataframes, ignore_index=True)

# Display the first few rows of the combined dataframe
combined_df.head()
```

```python
# Specify the output file path
output_file = os.path.join('C:\\Users\\benfi\\Defensive_Players_NFL\\data', 'cfb_def_stats.csv')

# Save the combined dataframe to a new CSV file
combined_df.to_csv(output_file, index=False)

print(f'Combined DataFrame has been saved as "cfb_def_stats.csv"')
```
